##### Configure Parameters

##### Data Visualization

```python
# Configure parameters
sns.set(style = 'ticks')


# Create sublots
fig, axes = plt.subplots(1, 2, figsize=(15, 5.5))

#Custom Colors
color_mapping = {1: '#FF922D', 0: '#0853A6'}

#  Subplot 1
sns.barplot(data = data, x = y, y = 'LIMIT_BAL', hue = y, palette = color_mapping, ax = axes[0])

# Subplot 2
sns.histplot(data = data, x = 'LIMIT_BAL', hue = y, palette = color_mapping, ax = axes[1])


# Display
plt.suptitle('Limit Balance by Default Indicator')
plt.tight_layout()
plt.show()
```
