### Exporatory Data Analysis

##### Notebook Setup

```python
# Import libraries

import pandas as pd
import numpy as np
```

##### Data Pull

```python
df = pd.read_csv('../02_data/01_raw/credit_card_data.csv')
df.head()
```

##### Data Pre-Processing

```python
# Use first row as columns
df.columns = df.iloc[0].values

# Drop row containing column values
data = df.iloc[1:]

# Drop first column
data = data.drop(columns = 0)

# Change data types
for col in list(data.columns):
    if col != 'default payment next month':
        data[col] = data[col].astype('int')

# Rename target column
data.rename(columns = {'default payment next month':'DEFAULT_IND',
                       'PAY_0':'PAY_1'}, inplace = True)
```

##### Save

```python
data.to_csv('../02_data/02_processed/credit_card_data_processed.csv')
```
